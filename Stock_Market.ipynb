```python
#Stock Market Visualization

# Import yfinance package
import yfinance as yf
# Import matplotlib for plotting
import matplotlib.pyplot as plt
```

```python
#Reliance Industries Limited (RELIANCE.NS)

##FOR CLOSE RANGE OF DATES

# Set the start and end date
start_date = '2020-01-01'
end_date = '2020-05-12'

# Set the ticker
ticker = 'RELIANCE.NS'

# Get the data
data = yf.download(ticker, start_date, end_date)


# Plot adjusted close price data
data['Adj Close'].plot()
print('For Closer Range of Dates')
plt.show()

##FOR WIDER RANGE OF DATES

# Set the start and end date
start_date2 = '2012-01-01'
end_date2 = '2022-05-12'

# Get the data
data2 = yf.download(ticker, start_date2, end_date2)

# Plot adjusted close price data
data2['Adj Close'].plot()
print('For Wider Range of Dates')
plt.show()
```

## IMPACT OF MEDIA AND BIG PLAYERS

```python
#Tesla vs Microsoft
#Effect of Media and Big Players

import matplotlib.pyplot as plt

# Set the start and end date
start_date = '2020-01-01'
end_date = '2020-05-12'

# Set the ticker
ticker = 'TSLA'
ticker1 = 'MSFT'

# Get the data
data = yf.download(ticker, start_date, end_date)
data2 = yf.download(ticker1, start_date, end_date)

fig,ax1=plt.subplots()

color='tab:red'
ax1.set_xlabel('Date')
ax1.set_ylabel('Tesla',color=color)
ax1.plot(data['Adj Close'],color=color)
ax1.tick_params(axis='y',labelcolor=color)

#graph for 2 different x axis
ax2=ax1.twinx()

color='tab:blue'
ax2.set_ylabel('Microsoft',color=color)
ax2.plot(data2['Adj Close'],color=color)
ax2.tick_params(axis='y',labelcolor=color)

fig.tight_layout()

fig.autofmt_xdate()
```
